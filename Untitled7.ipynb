# Seaborn Library

# 1. Histogram

```python
import seaborn as sns
```

```python
# Sample Data
data = sns.load_dataset('iris')
```

```python
data.head(3)
```

```python
# create histogram
sns.histplot(data=data, x='sepal_length', kde = True)

# add label and title
plt.xlabel('sepal_length')
plt.ylabel('frequency')
plt.title("Histogram of sepal length")



plt.show()
```

```python
# Boxplot
```

```python
# sample data
data = sns.load_dataset('tips')
data.head(3)
```

```python
# box plot
sns.boxplot(data = data, x="day", y="total_bill", palette="Set2")

plt.xlabel('day')
plt.ylabel('total_bill')
plt.title("Box plot of total bill of day")




plt.show()
```

```python
# box plot
sns.violinplot(data = data, x="day", y="total_bill", palette="pastel")

plt.xlabel('day')
plt.ylabel('total_bill')
plt.title("Box plot of total bill of day")




plt.show()
```
